# Data Cleaning

```python
#check data types
df.info()
```

* The value in dollars is a string object, it need to be changed to float.

```python
#check for missing values
df.isna().sum()
```

* The marking column consist of 158 nulls. Further investigation required on it.

```python
#check unique values for marking column
df['marking'].unique()
```

* The marking column have two unique values(NaN and None),  The column is not very usefull, it is warranted to drop it.

```python
#drop the marking coloumn. It has NAN and none, not usable at all
df =df.drop('marking', axis=1)
```

```python
#confirm that marking column is dropped
# the value is in string format
df.head()
```

* The value column is in string format, this will be changed to float

```python
#converting string for value to float
#remove dollar $ sign
df['value'] = df['value'].str.replace('$', '')
#remove the .
df['value'] = df['value'].str.replace('.', '')
#change value to be  float
df['value in dollars']=df['value'].astype('float')
```

```python
#drop the old value column. Newly created one will be used. 
df = df.drop('value', axis =1)
```

```python
# check new dataframe
df.head()
```

```python
#Create a function: This is to create a subregion using the above dictionary
def get_subregion(country):
    ''' get_subregion(country)
    
        The function to loop throgh a dictionary in order to create a new sub-region
        a sub-region comprises of countries in same regions of the world.
        
        "country" - function takes one argument, which is name of country'''
    
    for a,i in Sub_region.items():
        if country in i:
            return(a)
            break
```

```python
#Apply the subregion function to create a new column named region 
df["region"] = df["country"].apply(get_subregion)
```

## Further Feature Engineering
#### Create a new column to have a continental view, caribbean countries are treated seperately
#### The Continents are Africa, Europ,Asia, Oceania,North and South America, plus the caribbean

```python
#Dictionary to group region into continets
Continent={'Africa':['Central Africa','North Africa', 'Southern Africa','West Africa','East Africa'],
           'Europe':['Eastern Europe','Northern Europe','Southern Europe','Western Europe'],
           'Asia':['Central Asia','Eastern Asia','South-Eastern Asia','Southern Asia','Western Asia'],
           'North America' : 'North America',
           'South America' : 'South America',
           'Caribbean' : 'Caribbean',
           'Oceania' :'Oceania'}
```

```python
#Create a function: This is to create a continent using the above dictionary
def get_Continent(region):
    ''' get_Continent(region)
    
        The function to loop throgh a dictionary in order to create a continent
        a continent comprises of sub_regions in same area of the world.
        
        "region" - function takes one argument, which is name of region'''
    
    for a,i in Continent.items():
        if region in i:
            return(a)
            break
```

```python
#Apply the subregion function to create a new column named continent 
df["continent"] = df["region"].apply(get_Continent)
```
